```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("Titanic-Dataset.csv")
Df=pd.DataFrame(df)
Df.tail()
```

```python
Df.head()
```

```python
Df.describe()
```

```python
Df.describe(include=['O']) #object datatype summarization
```

```python
nf = df.select_dtypes(include=[np.number]).columns.tolist()
```

```python
df[nf].hist(figsize=(12, 10), bins=20)
plt.suptitle("Histograms of Numeric Features", fontsize=16)
plt.show()
```

```python
plt.figure(figsize=(12, 8))
for i, col in enumerate(nf, 1):
    plt.subplot(3, 3, i)
    sns.boxplot(y=df[col])
    plt.title(f"Boxplot of {col}")
plt.tight_layout()
plt.show()
```

```python
sns.pairplot(df[["Survived", "Age", "Fare", "Pclass"]], hue="Survived")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap="coolwarm", center=0)
plt.title("Correlation Matrix")
plt.show()
```

```python
sns.countplot(x="Survived", data=df)
plt.title("Survival Distribution")
plt.show()
```

```python
sns.countplot(x="Sex", hue="Survived", data=Df)
plt.title("Survival by Gender")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.kdeplot(Df[Df["Survived"] == 1]["Age"], shade=True, label="Survived")
sns.kdeplot(Df[Df["Survived"] == 0]["Age"], shade=True, label="Not Survived")
plt.title("Age Distribution by Survival")
plt.legend()
plt.show()
```
